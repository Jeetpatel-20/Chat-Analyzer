```python
import pandas as pd
import re
```

```python
f = open('WhatsApp Chat with Idiots.txt', 'r', encoding = 'utf-8') #24 hour format
g = open('WhatsApp Chat with Idiots 12hr.txt', 'r', encoding = 'utf-8') #12 hour format
```

```python
data_24 = f.read()
data_12 = g.read()
# read the data of the txt file
```

```python
print(data_24)
```

```python
print(data_12)
```

```python
pattern1 = '\d{2}/\d{2}/\d{2}, \d{2}:\d{2} - '
pattern2 = '\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}\s*[ap]m - '
```

```python
messages_24 = re.split(pattern1, data_24)[1:] # for 24 hour format
```

```python
messages_12 = re.split(pattern2, data_12) # for 12 hour format
```

```python
messages_24
```

```python
messages_12
```

```python
dates = re.findall(pattern1, data_24)
dates
```

Creating DataFrame

```python
df = pd.DataFrame({'user_message':messages_24, 'Dates':dates})
df['Date'] = pd.to_datetime(df['Dates'], format = '%d/%m/%y, %H:%M - ')
df.drop(columns = ['Dates'], inplace = True)
```

Separating user and messages

```python
users = []
messages = []
for message in df['user_message']:
  entry = re.split('(\w+?):\s', message)
  if entry[1:]:
    users.append(entry[1])
    messages.append(entry[2])
  else:
    users.append('group_notification')
    messages.append(entry[0])

df['user'] = users
df['message'] = messages
```

Deleting user_message column

```python
df.drop(columns = ['user_message'], inplace = True)
```

Extracting Year, Month, Day, Hour and mintes from the date column

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Month'] = df['Date'].dt.month_name()
```

```python
df['Day'] = df['Date'].dt.day
```

```python
df['Hour'] = df['Date'].dt.hour
```

```python
df['Minute'] = df['Date'].dt.minute
```
